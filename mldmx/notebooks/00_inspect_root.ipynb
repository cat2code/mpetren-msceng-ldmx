```python
import uproot

path = "../src/mldmx/data/events.root"
f = uproot.open(path)

# list TTrees at top-level
trees = []
for k, obj in f.items():
    if hasattr(obj, "num_entries") and hasattr(obj, "keys"):
        trees.append((k, obj.num_entries))
trees
```

```python
tree_name = trees[0][0]  # or set manually
t = f[tree_name]

branches = list(t.keys())
len(branches), branches[:30]
```

```python
[b for b in branches if "Ecal" in b or "ecal" in b][:80]
```

```python
import awkward as ak

some = [b for b in branches if "Ecal" in b][:10]  # adjust
arrs = t.arrays(some, entry_stop=5, library="ak")

for b in some:
    a = arrs[b]
    if ak.is_list_type(a.type):
        print(b, "jagged", "lens:", ak.to_list(ak.num(a, axis=1)))
    else:
        print(b, "flat", ak.to_list(a))
```
